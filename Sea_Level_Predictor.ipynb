```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import linregress

df= pd.read_csv("epa-sea-level.csv")
df = df.dropna(subset=["Year", "CSIRO Adjusted Sea Level"])
x = df["Year"]
y = df['CSIRO Adjusted Sea Level']
fig , ax = plt.subplots(figsize = (12,6))
ax= plt.scatter(x,y,label = "Origional")
df = df.dropna(subset=["Year", "CSIRO Adjusted Sea Level"])
# Best Fit line 
reg = linregress(x, y)
x_forecast = pd.Series([i for i in range(1850 , 2051)])
y_forecast = reg.slope* x_forecast + reg.intercept 
plt.plot(x_forecast , y_forecast , "r--" , label = "1st Best Fit Line")

df_forc = df.loc[df["Year"]>=2000 ]
new_x = df_forc["Year"]
new_y = df_forc["CSIRO Adjusted Sea Level"]
# 2nd best fit line 

reg_new = linregress(new_x,new_y)
x_newforcast = pd.Series([i for i in range (2000 , 2051)])
y_newforcast = reg_new.slope*x_newforcast + reg_new.intercept
plt.plot(x_newforcast , y_newforcast , "green" ,label = "2nd Best Fit Line" )
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.grid(True)
fig.savefig("plot.png")






```
